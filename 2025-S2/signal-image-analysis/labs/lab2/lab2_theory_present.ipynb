```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Временная ось
t = np.linspace(0, 2 * np.pi, 200)

# Сигналы: один синус, другой - сдвинутый
x = np.sin(t)
y = np.sin(t + np.pi / 4)  # сдвиг на 45 градусов

fig, ax = plt.subplots(figsize=(10, 3))
ax.plot(t, x, label="x(t)")
ax.plot(t, y, linestyle="--", label="y(t)")
ax.set_title("Сигналы x(t) и y(t)")
ax.set_xlabel("Время")
ax.set_ylabel("Амплитуда")
ax.legend()
ax.grid(True)
plt.show()
```

```python
# Автокорреляция сигнала x
auto_x = np.correlate(x, x, mode="full")
lags_x = np.arange(-len(x)+1, len(x))

fig, ax = plt.subplots(2, 1, figsize=(10, 6))

# Сигнал
ax[0].plot(t, x, label="x(t)")
ax[0].set_title("Сигнал x(t)")
ax[0].set_xlabel("Время")
ax[0].set_ylabel("Амплитуда")
ax[0].grid(True)
ax[0].legend()

# Автокорреляция
ax[1].plot(lags_x, auto_x)
ax[1].set_title("Автокорреляция x(t)")
ax[1].set_xlabel("Сдвиг (lag)")
ax[1].set_ylabel("Корреляция")
ax[1].grid(True)

fig.tight_layout()
plt.show()
```

```python
# Взаимная корреляция x и y
cross_xy = np.correlate(x, y, mode="full")
lags_xy = np.arange(-len(x)+1, len(x))

fig, ax = plt.subplots(2, 1, figsize=(10, 6))

# Два сигнала
ax[0].plot(t, x, label="x(t)")
ax[0].plot(t, y, linestyle="--", label="y(t)")
ax[0].set_title("Сигналы x(t) и y(t)")
ax[0].set_xlabel("Время")
ax[0].set_ylabel("Амплитуда")
ax[0].legend()
ax[0].grid(True)

# Взаимная корреляция
ax[1].plot(lags_xy, cross_xy, color="orange")
ax[1].set_title("Взаимная корреляция x и y")
ax[1].set_xlabel("Сдвиг (lag)")
ax[1].set_ylabel("Корреляция")
ax[1].grid(True)

fig.tight_layout()
plt.show()
```
