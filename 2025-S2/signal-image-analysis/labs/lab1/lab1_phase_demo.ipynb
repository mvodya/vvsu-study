```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 2*np.pi, 500)
```

```python
y1 = np.sin(t)              # фаза 0
y2 = np.sin(t + np.pi/4)    # фаза +π/4
y3 = np.sin(t + np.pi/2)    # фаза +π/2

fig, ax = plt.subplots(figsize=(8,4))
ax.plot(t, y1, label="φ = 0")
ax.plot(t, y2, label="φ = π/4")
ax.plot(t, y3, label="φ = π/2")

ax.set_title("Влияние фазы на сигнал")
ax.set_xlabel("Время")
ax.set_ylabel("Амплитуда")
ax.legend()
ax.grid(True)
plt.show()
```
