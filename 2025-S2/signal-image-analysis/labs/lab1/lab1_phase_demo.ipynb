```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 2*np.pi, 500)
```

```python
y1 = np.sin(t)              #  0
y2 = np.sin(t + np.pi/4)    #  +/4
y3 = np.sin(t + np.pi/2)    #  +/2

fig, ax = plt.subplots(figsize=(8,4))
ax.plot(t, y1, label=" = 0")
ax.plot(t, y2, label=" = /4")
ax.plot(t, y3, label=" = /2")

ax.set_title("   ")
ax.set_xlabel("")
ax.set_ylabel("")
ax.legend()
ax.grid(True)
plt.show()
```
