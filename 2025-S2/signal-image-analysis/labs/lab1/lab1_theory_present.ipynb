```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 1, 500)

# Аналоговый сигнал (непрерывная синусоида)
x = np.sin(2*np.pi*5*t)

# Дискретизация (берем выборки)
samples_t = np.linspace(0, 1, 20)
samples_x = np.sin(2 * np.pi * 5 * samples_t)

fig, ax = plt.subplots(figsize=(10, 3))

# Аналоговый сигнал
ax.plot(t, x, label="Аналоговый сигнал")

# Дискретные отсчеты
ax.stem(samples_t, samples_x, linefmt="r-", markerfmt="ro", basefmt=" ", label="Дискретный сигнал")

ax.set_title("Сигнал: аналоговый и его дискретные отсчёты")
ax.set_xlabel("Время (с)")
ax.set_ylabel("Амплитуда")
ax.legend()
ax.grid(True)

plt.show()
```

```python
# Временная ось
t = np.linspace(0, 1, 500)

# Сигнал - сумма двух синусоид
x = np.sin(2 * np.pi * 5 * t) + 0.5 * np.sin(2 * np.pi * 12 * t)

# Сигнал во времени
fig, ax = plt.subplots(figsize=(10, 3))
ax.plot(t, x)
ax.set_title("Сигнал во времени (сумма 5 Гц и 12 Гц)")
ax.set_xlabel("Время (сек)")
ax.set_ylabel("Амплитуда")
ax.grid(True)
plt.show()

# Преобразование Фурье
N = len(x)
X = np.fft.fft(x)  # БПФ
freqs = np.fft.fftfreq(N, d=t[1]-t[0])  # ось частот

# Берем только положительные частоты
mask = freqs >= 0
freqs = freqs[mask]
amplitude = np.abs(X[mask]) / N * 2  # нормирование

fig, ax = plt.subplots(figsize=(10, 3))
ax.stem(freqs, amplitude, basefmt=" ")
ax.set_title("Амплитудный спектр")
ax.set_xlabel("Частота (Гц)")
ax.set_ylabel("Амплитуда")
ax.set_xlim(0, 20)
ax.grid(True)
plt.show()
```

```python
t = np.linspace(0, 2 * np.pi, 500)

# Одна и та же частота, разные фазы
y1 = np.sin(t)               # фаза 0
y2 = np.sin(t + np.pi/4)     # фаза +π/4
y3 = np.sin(t + np.pi/2)     # фаза +π/2

fig, ax = plt.subplots(figsize=(10,4))
ax.plot(t, y1, label="φ = 0")
ax.plot(t, y2, label="φ = π/4")
ax.plot(t, y3, label="φ = π/2")
ax.set_title("Влияние фазы на синусоиду")
ax.set_xlabel("Время")
ax.set_ylabel("Амплитуда")
ax.legend()
ax.grid(True)
plt.show()
```
