```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 1, 500)

#   ( )
x = np.sin(2*np.pi*5*t)

#  ( )
samples_t = np.linspace(0, 1, 20)
samples_x = np.sin(2 * np.pi * 5 * samples_t)

fig, ax = plt.subplots(figsize=(10, 3))

#  
ax.plot(t, x, label=" ")

#  
ax.stem(samples_t, samples_x, linefmt="r-", markerfmt="ro", basefmt=" ", label=" ")

ax.set_title(":     ")
ax.set_xlabel(" ()")
ax.set_ylabel("")
ax.legend()
ax.grid(True)

plt.show()
```

```python
#  
t = np.linspace(0, 1, 500)

#  -   
x = np.sin(2 * np.pi * 5 * t) + 0.5 * np.sin(2 * np.pi * 12 * t)

#   
fig, ax = plt.subplots(figsize=(10, 3))
ax.plot(t, x)
ax.set_title("   ( 5   12 )")
ax.set_xlabel(" ()")
ax.set_ylabel("")
ax.grid(True)
plt.show()

#  
N = len(x)
X = np.fft.fft(x)  # 
freqs = np.fft.fftfreq(N, d=t[1]-t[0])  #  

#    
mask = freqs >= 0
freqs = freqs[mask]
amplitude = np.abs(X[mask]) / N * 2  # 

fig, ax = plt.subplots(figsize=(10, 3))
ax.stem(freqs, amplitude, basefmt=" ")
ax.set_title(" ")
ax.set_xlabel(" ()")
ax.set_ylabel("")
ax.set_xlim(0, 20)
ax.grid(True)
plt.show()
```

```python
t = np.linspace(0, 2 * np.pi, 500)

#     ,  
y1 = np.sin(t)               #  0
y2 = np.sin(t + np.pi/4)     #  +/4
y3 = np.sin(t + np.pi/2)     #  +/2

fig, ax = plt.subplots(figsize=(10,4))
ax.plot(t, y1, label=" = 0")
ax.plot(t, y2, label=" = /4")
ax.plot(t, y3, label=" = /2")
ax.set_title("   ")
ax.set_xlabel("")
ax.set_ylabel("")
ax.legend()
ax.grid(True)
plt.show()
```
